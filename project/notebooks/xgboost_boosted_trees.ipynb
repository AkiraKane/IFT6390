```python
import pandas as pd
import numpy as np
import xgboost as xgb

train = pd.read_csv("../data/train.csv")
test = pd.read_csv("../data/test.csv")
```

```python
def preprocessing_data(data):
    processed_data = pd.get_dummies(data)
    processed_data = processed_data.fillna(processed_data[:data.shape[0]].mean())
    return processed_data
```

```python
preprocessed_data = preprocessing_data(train)

X = preprocessed_data.as_matrix()[:,:-1]
y = train.as_matrix()[:,-1]
```

```python
xgb_model = xgb.XGBRegressor().fit(X, y)
preds = xgb_model.predict(preprocessing_data(test).as_matrix())
```
